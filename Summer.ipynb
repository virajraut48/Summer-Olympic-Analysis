### Assignment 2 : Summer Olympics Data Analysis 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Summer.csv")
```

```python
df.head()
```

### 1. In how many cities summer olympics is held so far?

```python
number_of_cities=len(df['City'].unique())
number_of_cities
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = []

for sport in df['Sport'].unique():
    data.append([sport , len(df[(df['Sport'] == sport)&(df['Medal'] == 'Gold')])])

plot_data = pd.DataFrame(data,columns = ['Sport','Number of Gold Medals']).sort_values(by='Number of Gold Medals', ascending=False).head()
print(plot_data)
plot_data.plot(x = 'Sport', y = 'Number of Gold Medals', kind = 'bar', title='Top-5 sports having most number Gold Medals',grid=True,figsize = (10,5), color = "orange")
```

### 3. Which Sport having most number of medals so far?

```python
data = []

for sport in df['Sport'].unique():
    data.append([sport , len(df[(df['Sport'] == sport)])])

plot_data = pd.DataFrame(data,columns = ['Sport','No. of medals']).sort_values(by='No. of medals', ascending=False).head()
print(plot_data)
plot_data.plot(x = 'Sport', y = 'No. of medals', kind = 'bar',title='Top-5 sports having most number of Medals',grid=True, figsize = (10,5), color = 'DarkCyan')
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []

for athlete in df['Athlete'].unique():
    data.append([athlete , len(df[(df['Athlete'] == athlete)])])

plot_data = pd.DataFrame(data,columns = ['Athlete','No. of medals']).sort_values(by='No. of medals', ascending=False).head()
print(plot_data)
plot_data.plot(x = 'Athlete', y = 'No. of medals', kind = 'bar', title='Top-5 Players having most number of Medals',grid=True,figsize = (10,5), color = 'Pink')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data = []

for athlete in df['Athlete'].unique():
    data.append([athlete , len(df[(df['Athlete'] == athlete) & (df['Medal'] == "Gold")])])

plot_data = pd.DataFrame(data,columns = ['Athlete','No. of gold medals']).sort_values(by='No. of gold medals', ascending=False).head()
print(plot_data)
plot_data.plot(x = 'Athlete', y = 'No. of gold medals', kind = 'bar', title='Top-5 Players having most number of Gold Medals',grid=True,figsize = (10,5), color = 'DarkGreen')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
india_first_gold_year=df[(df.Country =='IND') & (df.Medal =='Gold')][['Year','Medal']]
india_first_gold_year=india_first_gold_year.groupby('Year').count().reset_index()
india_first_gold_year=india_first_gold_year.sort_values(by='Year',ascending='True').loc[0,'Year']
india_first_gold_year
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []
for event in df['Event'].unique():
    data.append([event , len(df[(df['Event'] == event)])])

plot_data = pd.DataFrame(data,columns = ['Event','Athletes']).sort_values(by='Athletes', ascending=False).head()
print(plot_data)
plot_data.plot(x = 'Event', y = 'Athletes', kind = 'bar',title='Top-5 most popular events in terms on number of players',grid=True, figsize = (10,5), color = 'Red')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data = []
for sport in df['Sport'].unique():
    data.append([sport , len(df[(df['Sport'] == sport) & (df['Medal'] == "Gold") & (df['Gender'] == "Women")])])

plot_data = pd.DataFrame(data,columns = ['Sport','No of female athlete']).sort_values(by='No of female athlete', ascending=False).head()
print(plot_data)
plot_data.plot(x = 'Sport', y = 'No of female athlete',title='Top-5 Sport that having most female Gold Medalists',grid=True, kind = 'bar', figsize = (10,5), color = "Yellow")
```
